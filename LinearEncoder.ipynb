```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline

from py.PY_LinearAutoEncoder import LinearAutoEncoder
```

```python
size = 64
n = size * size
def readImg(path):
    img = cv.imread(path, 0)
    
    img = cv.resize(img, (size, size))
    x = np.reshape(img, (1,n))
    x = x.astype(float)
    x /= 255
    return x
```

```python
kazuha1 = readImg("statics/kazuha2.jpg")
encoder = LinearAutoEncoder(n, 5)
encoder.train(kazuha1, 30000, 0.00017)
encode, decode = encoder.forward(kazuha1)
losses = encoder.getLoss()
```

```python
plt.figure(figsize=(3,10.5))  # 调整画布大小
plt.subplot(3,1,1)
plt.imshow(decode.reshape(size, size))
plt.title('Test Kazuha')
plt.subplot(3,1,2)
plt.imshow(kazuha1.reshape(size,size))
plt.title('Original Kazuha')
plt.subplot(3,1,3)
plt.plot(losses[10:])
plt.title('Loss')
plt.show()
```
