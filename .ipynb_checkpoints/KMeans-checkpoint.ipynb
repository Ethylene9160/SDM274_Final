```python
import numpy as np
from scipy.spatial.distance import cdist
class KMeans:
    def __init__(self, n_clusters=3, max_iter=300, tol=1e-4):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tol = tol
        self.centroids = None
    
    def fit(self, X):
        # 初始化质心，随机选取输入样本。
        self.centroids = X[np.random.choice(X.shape[0], self.n_clusters, replace=False)]
        for i in range(self.max_iter):
            # 分配样本到最近的质心
            distances = cdist(X, self.centroids, 'euclidean')
            labels = np.argmin(distances, axis=1)
            
            # 计算新的质心
            new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(self.n_clusters)])
            
            # 检查质心是否变化很小
            if np.all(np.abs(new_centroids - self.centroids) <= self.tol):
                break
            
            self.centroids = new_centroids
        
        self.labels_ = labels
        return self
    
    def predict(self, X):
        distances = cdist(X, self.centroids, 'euclidean')
        return np.argmin(distances, axis=1)

class SoftKMeans(KMeans):
    def __init__(self, n_clusters=3, max_iter=300, tol=1e-4, beta=1.0):
        super().__init__(n_clusters, max_iter, tol)
        self.beta = beta
    
    def fit(self, X):
        # 初始化质心
        self.centroids = X[np.random.choice(X.shape[0], self.n_clusters, replace=False)]
        
        for i in range(self.max_iter):
            # 计算样本和质心之间的距离
            distances = cdist(X, self.centroids, 'euclidean')
            
            # 计算责任（soft labels）
            exp_neg_beta_d = np.exp(-self.beta * distances)
            weights = exp_neg_beta_d / np.sum(exp_neg_beta_d, axis=1, keepdims=True)
            
            # 计算新的质心
            new_centroids = np.dot(weights.T, X) / np.sum(weights.T, axis=1, keepdims=True)
            
            # 检查质心是否变化很小
            if np.all(np.abs(new_centroids - self.centroids) <= self.tol):
                break
            
            self.centroids = new_centroids
        
        self.labels_ = np.argmax(weights, axis=1)
        return self
```

```python
   # 定义一个函数来读取小麦数据集的txt文件
import pandas as pd
def read_seeds_dataset(file_path):
    try:
        # 使用pandas读取txt文件
        data = pd.read_csv(file_path, sep='\s+', header=None)
        # 给数据集的列命名
        data.columns = ['area', 'perimeter', 'compactness', 'length_kernel', 'width_kernel',
                        'asymmetry_coefficient', 'length_kernel_groove', 'class']
        return data
    except Exception as e:
        return str(e)  # 如果出现错误，返回错误信息
    
seeds_data = read_seeds_dataset('seeds_dataset.txt')
seeds_data = [
    [0.1,0.1],
    [0.1,0.2],
    [0.2,0.1],
    [0.2,0.2],
    
    [2.1,0.1],
    [2.1,0.2],
    [2.2,0.1],
    [2.2,0.2],
    
    [0.1,2.1],
    [0.1,2.2],
    [0.2,2.1],
    [0.2,2.2],
    
    [2.1,2.1],
    [2.1,2.2],
    [2.2,2.1],
    [2.2,2.2],
             ]
print(seeds_data)
# 创建KMeans和SoftKMeans实例
seeds_data = np.array(seeds_data)
kmeans = KMeans(n_clusters=4)
soft_kmeans = SoftKMeans(n_clusters=4,beta=0.5)  # beta参数控制soft assignment的程度

# 拟合小麦数据集
# kmeans.fit(seeds_data.iloc[:, :-1].values)
# soft_kmeans.fit(seeds_data.iloc[:, :-1].values)

kmeans.fit(seeds_data)
soft_kmeans.fit(seeds_data)

# 得到KMeans和SoftKMeans的聚类标签
kmeans_labels = kmeans.labels_
soft_kmeans_labels = soft_kmeans.labels_

# 显示第一个算法的质心和标签，以确认实现正确
(kmeans.centroids, kmeans_labels)
```

```python
class PCA:
    def __init__(self):
        #self.k = k
        self.hasTrained = False
        self.mean = np.array([[]])
        self.transformedMatrix = np.array([[]])
        
    def train(self, data):
        #X=data.copy # copy, prevent to influence the data outside.
        self.mean = np.mean(data, axis=0)
        X = data-self.mean
        
        # cov:
        cov_mat = np.cov(X.T)
        
        # eigens
        eigenvalues, eigenvectors = np.linalg.eig(cov_mat)
        
        #sort eigenvector
        idx = np.argsort(eigenvalues)[::-1]
        
        # store transform matrix
        self.transformedMatrix = eigenvectors[:, idx]
        self.hasTrained = True
    
    def transform(self, data, k):
        if self.hasTrained == False:
            print('has not trained!')
            self.train(data)
        transformk = self.transformedMatrix[:, :k]
        # X = data.copy()
        X = data-self.mean
        return np.dot(X, transformk)
```
